Pour gérer le déséquilibre des classes dans la variable cible **SeriousDlqin2yrs**, voici quelques approches possibles :

### 1. **Sous-échantillonnage (Under-sampling)** :
   - Réduire le nombre d'exemples de la classe majoritaire pour équilibrer les classes.
   - Cela peut entraîner une perte d'information, mais est utile pour des datasets très déséquilibrés.

### 2. **Sur-échantillonnage (Over-sampling)** :
   - Augmenter artificiellement le nombre d'exemples de la classe minoritaire en dupliquant les données ou en utilisant des techniques comme **SMOTE** (Synthetic Minority Over-sampling Technique).
   - Cela permet de conserver toutes les données de la classe majoritaire.

### 3. **Pondération des classes** :
   - Utiliser des poids pour donner plus d'importance à la classe minoritaire lors de la modélisation.
   - Cela est souvent intégré dans les algorithmes de machine learning.

### 4. **Création de nouvelles données** :
   - Générer des exemples synthétiques pour la classe minoritaire en utilisant des techniques avancées comme **GANs** (Generative Adversarial Networks).

### 5. **Stratification** :
   - Lors de la division des données en ensembles d'entraînement et de test, utiliser une stratification pour garantir que les proportions des classes sont respectées.

Appliquer le sur-échantillonnage pour équilibrer les classes dans le dataset.

```python
import pandas as pd
from sklearn.utils import resample
from io import StringIO

# Charger les données depuis le contenu fourni

data = pd.read_csv("data/cs-training.csv")

# Remplir les valeurs manquantes par la médiane pour éviter les problèmes
data['MonthlyIncome'].fillna(data['MonthlyIncome'].median(), inplace=True)
data['NumberOfDependents'].fillna(0, inplace=True)

# Séparer les classes majoritaires et minoritaires
majority_class = data[data['SeriousDlqin2yrs'] == 0]
minority_class = data[data['SeriousDlqin2yrs'] == 1]

# Sur-échantillonnage de la classe minoritaire
minority_class_oversampled = resample(minority_class,
                                      replace=True,
                                      n_samples=len(majority_class),
                                      random_state=42)

# Combiner les classes équilibrées
data_balanced = pd.concat([majority_class, minority_class_oversampled])

# Résultat
data_balanced_distribution = data_balanced['SeriousDlqin2yrs'].value_counts()
data_balanced_distribution
```

### 4. **Conclusion**
